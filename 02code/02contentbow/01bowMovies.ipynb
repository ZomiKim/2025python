```python
# movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: 참 if 조건 else 거짓 )

# test_list = movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: sorted(i['name'] for i in x ) if isinstance(x, list) else [] )


movies['str_genres'] = \
movies['genres'].fillna('[]').apply(literal_eval)\
    .apply(lambda x: sorted(i['name'] for i in x ) if isinstance(x, list) else [] )\
    .apply(lambda x : " ".join(x) if len(x)>0 else None)


```

```python
movies.head()
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
```

```python
movies.info()
```

```python
movies['year'] = movies['release_date'].dt.year
movies[['release_date', 'year']]
```

```python
# movies.isnull().sum() 
# movies.shape

(movies.isnull().sum() / movies.shape[0] )
```

```python
movies = movies.dropna()
movies.reset_index(drop=True)
```

```python
movies.isnull().sum()
```

```python
## bow기반 content based filltering

from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_vecter = CountVectorizer()
```

```python
genre_mat = bow_vecter.fit_transform(movies['str_genres'])
```

```python
bow_vecter.vocabulary_
```

```python
genre_mat.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_of_genre = cosine_similarity(genre_mat, genre_mat)
```
