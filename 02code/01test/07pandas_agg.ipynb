```python
import pandas as pd
```

```python
data = {
    "class": ["A", "A", "B", "B"],
    "name": ["철수", "영희", "민수", "지수"],
    "score": [80, 90, 70, 85]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby('class')

for key,value in df.groupby('class'):
    print(key)
    print(value)
```

```python
print(df.groupby('class')['score'].mean())
```

```python
data1 = {
    '학급': ['A', 'A', 'B', 'B', 'C', 'C'],
    '이름': ['철수', '영희', '민수', '수진', '지훈', '혜진'],
    '수학': [90, 80, 70, 85, 60, 95],
    '영어': [85, 88, 75, 70, 95, 90]
}

df1 = pd.DataFrame(data1)
df1
```

```python
df1_mean = df1.groupby('학급')[['수학', '영어']].mean().reset_index()
df1_mean
```

```python
df2_mean = df1.groupby('학급')[['수학','영어']].mean().reset_index()
print(df2_mean)
print(df2_mean.shape)
```

```python
df1.groupby('학급')['수학'].agg(['mean','max','count'])
```

```python
df1['수학'].mean()
```

```python
df1.groupby('학급')['수학'].describe()
```

```python
df1.groupby('학급')['수학'].agg(
    평균='mean',
    최고='max'
)

```

```python

df1.groupby('학급').agg(
    수학평균=('수학', 'mean'),
    영어평균=('영어', 'mean')
)
```
