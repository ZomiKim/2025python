```python
import random

def rand_add():
    a = random.randint(1,100)
    b = random.randint(1,100)
    answer = int(input(f'{a} + {b} = '))
    if answer == a+b:
        print("o")
    else:
        print("x")


rand_add()
```

### 리스트내포 / 리스트 컴프리헨션
- [표현식 for 변수 in]
- [표현식 for 변수 in 반복 if 조건] 
- [참 if 조건 else 거짓 for 변수 in 반복]  


```python
# 리스트내포/ 리스트 컴프리헨션
a = [ i for i in range(1,10,2)]
a

```

```python
b = [i for i in range(1,21) if i%2 == 0]
b
```

```python
c = [x for x in [-2,-1,0,2,5] if x > 0]
c
```

```python
score = 80
result = "pass" if score > 60 else "fail"
print(result)
```

```python
nums = [95, 60, 70, 40]
d = ["pass" if s>=70 else "fail" for s in nums]
d
```

```python
labels = ["even" if i % 2 == 0 else "odd" for i in range(1, 11)]
labels
```

```python
names = ["han", "kim", 'lee']
upper = [n.upper() for n in names]
upper
```

```python
def sum(n:int):
    total = 0
    for a in range(1,n+1) : 
        total += a
    return total
print(sum(5))    
print(sum(10000))    
```

```python
def sum1(n:int):
    total = 0 
    a=1
    while a <= n :
        total += a
        a += 1
    return total
print(sum1(5))
```

```python
price = [1200, 3000, 4500, 800]

for i in range (len(price)):
    price[i] = int(price[i] * 1.05)

price
```
