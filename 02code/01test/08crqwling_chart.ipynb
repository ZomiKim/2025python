```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.melon.com/chart/'
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7'
}

res = requests.get(url,headers=headers)
soup = BeautifulSoup(res.text,"html.parser")

title_one = soup.select_one("#tb_list .lst50 .ellipsis.rank01 a")
# print(title_one.text)


top_50 = soup.select("#lst50 > td > div")
# print(top_50)
# print(len(top_50))

for i in top_50:
    row = i
    rank = row.select_one('span.rank' )
    # print(rank)


    if rank is not None :
        print(rank.text)

    title = row.select_one('.ellipsis.rank01 > span > a')
    if title is not None :
        print(title.text, end='')

    singer = row.select_one('.ellipsis.rank01 > span > a')
    if title is not None :
        print(singer.text, end='')



```
