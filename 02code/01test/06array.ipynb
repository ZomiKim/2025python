```python
#동물별 평균 체온과 최대 체중 구하기
import numpy as np

#체온, 심박수, 체중(kg)
data = np.array([
    [38.5, 80, 35], #dog
    [39.1, 90, 50], #cat
    [36.8, 60, 300], #holes

])

ani = np.array(["dog","cat","holes"])

#체온, 심박수, 체중의 평균값
# avg_value = data.mean(axis=1)
avg_value = data.min(axis=0)
print(avg_value)

weight = data[:,2]
print(weight)

max_weight = data[:,2].max()
print(max_weight)

max_ani = ani[data[:, 2].argmax()]
print(max_ani)

print("동물별 평균 수치", avg_value.astype(int))
print("가장 무거운 동물", max_ani, f'{max_weight}kg')
```

```python
import numpy as np

# 곤충 [날개 수, 다리 수, 몸 길이(cm)]
data = np.array([
    [2, 6, 1.2],
    [4, 6, 3.4],
    [0, 6, 0.8],
    [0, 8, 2.0]
])

insects = np.array(["나비", "잠자리", "개미", "거미"])

# 다리가 6개인 곤충 추출
legs = data[data[:, 1] == 6]
print(legs)

#평균 몸 길이
mean_length = data[:,2].mean()
print(mean_length)
print(f'곤충 평균 몸 길이 : {round(mean_length,2)} cm' )

#날개 없는 동물
no_wings = insects[data[:,0] == 0]
print(no_wings)
```
