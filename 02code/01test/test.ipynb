````python
# 기본 입장료
base_price = 5000

# 나이 입력
age = int(input("나이를 입력하세요: "))

# 할인율 적용
if age < 8:
    price = base_price * 0.5
elif age < 18:
    price = base_price * 0.7
elif age >= 65:
    price = base_price * 0.6
else:
    price = base_price

# 결과 출력
print("입장료는", int(price), "원입니다.")
````
